### UI demo - gradio

```python
import numpy as np
import pandas as pd
from tensorflow.keras.models import load_model

# load the mobile phone price model
model = load_model("mobilephonemodel.keras")
```

```python
# pip install gradio
import gradio as gr

# https://www.gradio.app/guides/quickstart

# category names for predictions
categories = ['1: Cheap', '2: Avg-', '3: Avg+', '4: Expensive']


def predict_phone(battery_power, fc, pc, int_memory, n_cores, ram, height, width, sc_h, sc_w, talktime, weight):
    tester_row = {
        'battery_power': int(battery_power), 
        'fc': int(fc), 
        'int_memory': int(int_memory), 
        'mobile_wt': int(weight),
        'n_cores': int(n_cores),
        'pc': int(pc),
        'px_width': int(width),
        'px_height': int(height), 
        'ram': int(ram),
        'sc_h': int(sc_h), 
        'sc_w': int(sc_w), 
        'talk_time': int(talktime), 
    }

    categories = ['1: Cheap', '2: Avg-', '3: Avg+', '4: Expensive']

    tester_row = pd.DataFrame([tester_row])
    result = model.predict(tester_row)[0]
    result_text = categories[np.argmax(result)]
    
    return result_text

demo = gr.Interface(
    fn=predict_phone,
    inputs=["text", "text", "text", "text", "text", "text", "text", "text", "text", "text", "text", "text"],
    outputs=["text"],
)

demo.launch()
```
